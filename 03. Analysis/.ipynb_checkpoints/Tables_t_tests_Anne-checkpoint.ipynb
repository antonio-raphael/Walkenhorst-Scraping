```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Markup t-tests

```python
df_markup = pd.read_csv('Markup-Tests.csv')
df_markup.head()
```

```python
def add_stars(p):
    if p < 0.01: return '***'
    elif p < 0.05: return '**'
    elif p < 0.1: return '*'
    else: return ''

df_display = df_markup.copy()
df_display = df_display.sort_values(by='mean', ascending=False)
df_display['Mean Markup (%)'] = df_display.apply(lambda x: f"{x['mean']:.2f}{add_stars(x['Adjusted_P_Value'])}", axis=1)
df_display['t-stat'] = df_display['t_statistic'].apply(lambda x: f"{x:.2f}")
df_display['P-Value'] = df_display['Adjusted_P_Value'].apply(lambda x: f"{x:.2f}")

table_data = df_display[['Category', 'Mean Markup (%)', 't-stat', 'P-Value']]

fig, ax = plt.subplots(figsize=(11, 8 ))
ax.axis('off')

# Creating table
tbl = ax.table(
    cellText=table_data.values,
    colLabels=table_data.columns,
    loc='center',
    cellLoc='center')

# Applying Styling
tbl.auto_set_font_size(False)
tbl.set_fontsize(10.5)
tbl.scale(1.0, 1.8) # Adjusted vertical scaling for readability

# Removing all cell borders
for (row, col), cell in tbl.get_celld().items():
    cell.set_linewidth(0)
    if row == 0:
        cell.set_text_props(weight='bold')
    if row == 10:
        cell.set_text_props(color='red', weight='bold')

# Manual Lines 
ax.axhline(y=0.97, xmin=0.05, xmax=0.95, color='black', linewidth=1.5) # Top line
ax.axhline(y=0.94, xmin=0.05, xmax=0.95, color='black', linewidth=0.8) # Header sep
ax.axhline(y=0.08, xmin=0.05, xmax=0.95, color='black', linewidth=1.5) # Bottom line

# 5. Positioned Text
plt.title("Table 1: Markup Test Results", 
          fontsize=14, weight='bold', pad=0, loc='center')

plt.text(0.05, 0.03, "Note: * p<0.1; ** p<0.05; *** p<0.01.", 
         ha='left', fontsize=10, style='italic', transform=ax.transAxes)

plt.tight_layout()
plt.savefig('Markup_TestsTable.png', bbox_inches='tight', dpi=300)
```

## 2. Price difference t-tests

```python
df_price = pd.read_csv('Price-Per-Unit-Tests.csv')
df_price.head()
```

```python
df_display2 = df_price.copy()
df_display2 = df_display2.sort_values(by='mean_diff', ascending=True)
df_display2['Mean Unit Price \n Difference'] = df_display2.apply(lambda x: f"{x['mean_diff']:.2f}{add_stars(x['Adjusted_P_Value'])}", axis=1)
df_display2['Mean Unit Price \n Walkenhorst'] = df_display2['estimate1'].apply(lambda x: f"{x:.2f}")
df_display2['Mean Unit Price \n Walmart'] = df_display2['estimate2'].apply(lambda x: f"{x:.2f}")
df_display2['t-statistic'] = df_display2['t_statistic'].apply(lambda x: f"{x:.2f}")
df_display2['P-Value'] = df_display2['Adjusted_P_Value'].apply(lambda x: f"{x:.2f}")
table_data = df_display2[['Category','Mean Unit Price \n Difference', 'Mean Unit Price \n Walkenhorst', 'Mean Unit Price \n Walmart', 't-statistic', 'P-Value']]

fig, ax = plt.subplots(figsize=(15, 8))
ax.axis('off')

tbl = ax.table(
    cellText=table_data.values,
    colLabels=table_data.columns,
    loc='center',
    cellLoc='center')

# Applying Styling
tbl.auto_set_font_size(False)
tbl.set_fontsize(9.5)
tbl.scale(1.0, 1.8) 

# Remove all cell borders
for (row, col), cell in tbl.get_celld().items():
    cell.set_linewidth(0)
    if row == 0:
        cell.set_text_props(weight='bold')
    if row == 10:
        cell.set_text_props(color='red', weight='bold')

# Manual Lines
ax.axhline(y=0.97, xmin=0.05, xmax=0.95, color='black', linewidth=1.5) 
ax.axhline(y=0.94, xmin=0.05, xmax=0.95, color='black', linewidth=0.8) 
ax.axhline(y=0.08, xmin=0.05, xmax=0.95, color='black', linewidth=1.5) 

# 5. Positioned Text
plt.title("Table 2: Unit Price Difference (Walmart - Walkenhorst) Test Results", 
          fontsize=14, weight='bold', pad=0, loc='center')

plt.text(0.05, 0.03, "Note: * p<0.1; ** p<0.05; *** p<0.01.", 
         ha='left', fontsize=10, style='italic', transform=ax.transAxes)

plt.tight_layout()
plt.savefig('UnitPrice_TestsTable.png', bbox_inches='tight', dpi=300)
```

## 3. Regression table

### a) Regression results (not adjusted for clustered standard errors)

```python
df_regression = pd.read_csv('Regression_Results_Not_Adjusted.csv')
df_regression.head()
```

```python
df_reg = df_regression.copy()
df_reg['Term'] = df_reg['Term'].str.replace('Walkenhorst_Category', '', regex=False)

df_reg['Estimate'] = df_reg.apply(lambda x: f"{x['Estimate']:.2f}{add_stars(x['p_value'])}", axis=1)
df_reg['Std. Error'] = df_reg['Std_Error'].apply(lambda x: f"({x:.2e})")
df_reg['t-value'] = df_reg['t_value'].apply(lambda x: f"{x:.2e}")
df_reg['P-Value'] = df_reg['p_value'].apply(lambda x: f"{x:.2e}")

table_data = df_reg[['Term', 'Estimate', 'Std. Error', 't-value', 'P-Value']]
table_data.columns = ['Variable', 'Coefficient', 'Std. Error', 't-statistic', 'P-Value']

# Plotting
fig, ax = plt.subplots(figsize=(16, 12)) 
ax.axis('off')

tbl = ax.table(
    cellText=table_data.values,
    colLabels=table_data.columns,
    bbox=[0.05, 0.15, 0.9, 0.7], 
    cellLoc='center')

tbl.auto_set_font_size(False)
tbl.set_fontsize(10)

for (row, col), cell in tbl.get_celld().items():
    cell.set_linewidth(0)
    if row == 0 :
        cell.set_text_props(weight='bold')
    if row == 2 and col == 1:
        cell.set_text_props(weight='bold')

num_rows = len(table_data)

# Horizontal Lines
ax.axhline(y=0.88, xmin=0.05, xmax=0.95, color='black', linewidth=1)
ax.axhline(y=0.89 , xmin=0.05, xmax=0.95, color='black', linewidth=1.5)
ax.axhline(y=0.13, xmin=0.05, xmax=0.95, color='black', linewidth=1.5)

# Positioned Text
ax.text(0.5, 0.92, "Table 3 : Unit Price ($/oz.) Regression Results", 
        ha='center', va='center', fontsize=16, weight='bold', transform=ax.transAxes)

ax.text(0.04, 0.09, "Note: * p<0.1; ** p<0.05; *** p<0.01. \nWalmart is the reference vendor (vendor = 1 for Walkenhorst, 0 for Walmart). \nAsian Food is the reference category.", 
        ha='left', va='center', fontsize=11, style='italic', transform=ax.transAxes)

plt.savefig('Regression_ResultsTable.png', bbox_inches='tight', dpi=300)
```
